## Libraries

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

from pathlib import Path
```

## Path

```python
img_dir = Path("../input/bengaliai-cv19/train_images/")
img_paths = list(img_dir.glob("*.png"))
img_paths[:5]
```

```python
idx = 5
print(str(img_paths[idx]))
img = cv2.imread(str(img_paths[idx]))
plt.imshow(img);
```

```python
plt.hist(img.reshape(-1), bins=50)
plt.show()
```

```python
img2 = img.copy()
img2[img > 150] = 255
img2[img <= 150] = 0
kernel = np.ones((3, 3), np.uint8)
img2 = cv2.morphologyEx(img2, cv2.MORPH_OPEN, kernel)
plt.imshow(img2);
```
