## Libraries

```python
import albumentations as A
import cv2
import matplotlib.pyplot as plt
import numpy as np

from pathlib import Path
```

## Path

```python
img_dir = Path("../input/bengaliai-cv19/train_images/")
img_paths = list(img_dir.glob("*.png"))
img_paths[:5]
```

```python
idx = 5
print(str(img_paths[idx]))
img = cv2.imread(str(img_paths[idx]))
plt.imshow(img);
```

```python
transforms = A.Compose([
    A.Cutout(num_holes=1, max_h_size=50, max_w_size=75, fill_value=255, always_apply=True)], 
    p=1.0)
```

```python
cutouted = transforms(image=img)["image"]
plt.imshow(cutouted);
```
